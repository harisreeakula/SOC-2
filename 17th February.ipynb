```python
# K means clusturing is an unsupervised learning algorithm that will attempt to group similar clusturs together from your data
# it is mainly used in clusturing similar documents,clusturing customers based on similar features
```

```python
from sklearn.datasets import make_blobs
```

```python
#the make_blobs functions is used to generate Synthetic DataSets for clusturing and classification tasks.this function will create clusters of data points with Gaussian distribution
```

```python
#creating random dataset
data=make_blobs(n_samples=200,n_features=2,centers=4,cluster_std=1.6,random_state=101)
# n_samples=total number of points equally divided among the clusters
# n_features=it indicates the number of features(columns)
# centers=number of clusters to be generated
#cluster_std=it sets the standard deviation of the clusters.High value makes the clusters to spread out
```

```python
data
```

```python
import matplotlib.pyplot as plt
```

```python
x,y=data
```

```python
plt.scatter(x[:,0],x[:,1],c=y,cmap='rainbow',edgecolor='black',s=250)
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.title("Scatter plot for k means ")
```

```python
data[0].shape
```

```python
from sklearn.cluster import KMeans
```

```python
Kmeans =KMeans(n_clusters=4)
```

```python
Kmeans.fit(data[0])
```

```python
Kmeans.cluster_centers_
```

```python
fig,(ax1,ax2)=plt.subplots(1,2,figsize=(10,6))
ax1.set_title('K means')
ax1.scatter(data[0][:,0],data[0][:,1],c=Kmeans.labels_,cmap="rainbow")
ax2.set_title("Original")
ax2.scatter(data[0][:,0],data[0][:,1],c=data[1],cmap="rainbow")
```

```python
from sklearn.model_selection import train_test_split
```

```python
x=df_feat
y=df['TARGET CLASS']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
KNN=KNeighborsClassifier(n_neighbors=3)
```

```python
KNN
```

```python
#to train model
KNN.fit(x_train,y_train)
```

```python
pred=KNN.predict(x_test)
```

```python
pred
```

```python
from sklearn.metrics import accuracy_score
```

```python
acc=accuracy_score(pred,y_test)
```

```python
acc
```

```python
# to find the error rate
error_rate=[]
for val in range(1,30):
    knn=KNeighborsClassifier(n_neighbors=val)
    knn.fit(x_train,y_train)   
    pred_i=knn.predict(x_test)    
    error_rate.append(np.mean(pred_i!=y_test))
```

```python
error_rate
```

```python
plt.figure(figsize=(6,3))
plt.plot(range(1,30),error_rate)
plt.title("Error rate vs value")
plt.xlabel("K")
plt.ylabel("Error Rate")
```
